```python
data_merge.head(10)
```

```python
data= data_merge.drop(['title','subject','date'],axis=1)
```

```python
def wordopt(text):
    text = str(text).lower() # Ensure text is string and lowercase
    text = re.sub(r'https?://\S+|www\.\S+', '', text) # Remove URLs
    text = re.sub(r'<.*?>+', '', text) # Remove HTML tags
    text = re.sub(r'[^a-z0-9\s]', '', text) # Keep only letters, numbers, and whitespace
    text = re.sub(r'\s+', ' ', text).strip() # Replace multiple spaces with a single space and strip whitespace
    return text if text else 'placeholder' # Return placeholder if text becomes empty
```

```python
data['text'] = data['text'].fillna('').astype(str).apply(wordopt)
# Ensure no empty strings remain after cleaning, replace with placeholder if necessary
data['text'] = data['text'].apply(lambda x: x if x and x.strip() else 'placeholder')
```

```python
x = data['text']
y = data['class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xv_train,y_train)
```
